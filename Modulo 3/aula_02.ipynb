## Analise de um dataset de compras de um e-commerce

```python
import pandas as pd

# Criar um dataframe
df = pd.read_csv('Ecommerce Purchases (1).csv')

# Ver apenas as primeiras linhas
display(df.head())
```

```python
#imprimir as colunas

#print(df.columns) 

for i in df.columns:
    print(i)
```

```python
# ver apenas uma coluna
print(df["Language"].head(2))

# ver mai que uma coluna
print(df[["Company","Language","CC Provider"]].head(5))
```

```python
# agrupar as vendas por idioma 
idioma = df.groupby("Language")["Language"].count()

print(idioma)

Job = df.groupby("Job")["Job"].count()

print(Job)
```

```python
# Somar uma coluna
total = df["Purchase Price"].sum()
print(f"Faturamento de vendas : $ {total:.2f} doláres")

# vendas por titulo de cartão usado
vendasCartao = df.groupby("CC Provider")["Purchase Price"].sum()
print(vendasCartao)

# apresente a nédia de vendas por idioma(mean)
MedianaCartao = df.groupby("Language")["Purchase Price"].mean()
print(MedianaCartao)
```

```python
# Criar filtros
# ver aepnas as vendas de um idioma especifico

el = df[df["Language"]=="el"]
display(el.head(2))
```

```python
# criar duas colunas 
# coluna 1- imposto de 10% sobre as vendas
df["imposto"] = df["Purchase Price"] * 0.1

display(df[["Purchase Price","imposto"]])

# coluna 2- calcular o valor da venda menos o imposto
# Valor_Líquido
df["Valor_Líquido"] = df["Purchase Price"] - df["imposto"]

display(df[["Purchase Price","imposto","Valor_Líquido"]])
```
